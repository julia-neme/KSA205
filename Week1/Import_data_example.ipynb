# Import data example

This example script takes you through importing data. There are may different ways to do this in Matlab depending on the type of data file you have. These are just a few examples.

We will use a library called `pandas` that is very useful for working with spreadsheet data (excel, csv, etc).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Method 1 

Use `pd.read_csv()` to read a csv file:

```python
dataRM = pd.read_csv('CTD_data_example.csv')
dataRM
```

## Method 2

Use `pd.read_excel()` to read a xlsx file:

```python
dataEX = pd.read_excel('CTD_data_example.xlsx')
dataEX
```

## Extract variables from the data

Let's get depths and salinities for plotting:

```python
depth = dataRM['Depth']
salinity = dataRM['SALINITY']
```

## Plot a profile
 
We can make a 'profile' plot of salinity versus depth:

```python
plt.plot(salinity, depth, color = 'k', marker = 'o');
```

Let's invert the y-axis so that greater depths are at the bottom and add appropriate labels:

```python
plt.plot(salinity, depth, color = 'k', marker = 'o');
plt.gca().invert_yaxis()
plt.xlabel('Salinity') # No units because this is sensor salinity, not absolute salinity
plt.ylabel('Depth (m)');
```

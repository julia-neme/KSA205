# Ekman transport

Ekman transport refers to movement of the ocean driven by a balance between wind stress and the Coriolis force. In this prac we will look at two examples of Ekman transport in the oceans: Upwelling in the Southern Ocean due to the influence of the Southern Hemisphere Westerly Winds, and coastal upwelling off the U.S. West Coast, which occurs seasonally in response to southward (AKA northerly) winds.  The Ekman equation states that the mass transport in the upper ocean as a result of a wind stress is proportional to the wind stress and inversely proportional to the Coriolis force:

$$
u = U . \rho_W = \frac{\tau^y}{f}
$$
$$
v = V . \rho_W = -\frac{\tau^x}{f}
$$

where u and v represent Ekman **mass** transport in the x and y directions, respectively, with units of **kg m-1 s-1**. U and V represent Ekman **volume** transport in the x and y directions, respectively, with units of **m3 m-1 s-1**. $\tau^x$ and $\tau^y$ represent the components of **wind stress** (units of **kg m-1 s-2**). $f$ is the Coriolis parameter as defined in the Geostrophic Currents prac (units of s-1), and $\rho_W$ is the density of water (1027 kg m-3 is a good estimate). Note how an Ekman transport in the x direction is driven by a wind stress in the y direction, and conversely and Ekman transport in the y direction is driven by a wind stress in the x direction. In other words, the transport is at 90 degrees to the wind direction.

# Part 1: Southern Ocean upwelling

In the Southern Ocean the Southern Hemisphere Westerly Winds (Westerlies) blow from West to East unimpeded by continents. These are some of the strongest winds over the ocean, famous for the ‘roaring forties’ and ‘furious fifties’ known to sailors. The Westerlies drive an Ekman transport to the left of the wind direction, that is, towards the North. Closer to Antarctica there are Easterly winds (blowing towards the west), and Ekman transport to the South, which ultimately sinks and flows to the north. The region of the Southern Ocean where waters are flowing north and south is known as the Antarctic Divergence:

<div align="center">
    <img src="upwelling.png" width="30%">
</div>

Water upwells from great depths here to replace the water moving away at the divergence, as explained in the lectures. This upwelling has important consequences for the ocean’s nutrient and carbon cycles, as it’s a very large transport of deep, carbon and nutrient rich water, to the ocean surface. This upwelled water is also relatively warm so it’s important for heat transfer and melting of sea-ice. Southern Ocean upwelling is a key part of the ocean’s global meridional overturning circulation, transforming deep waters back into surface waters.

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
import gsw
```

```python
data = pd.read_csv('monthly.2000.txt', skiprows = 42, header = None)[0]
```

Data looks like this: 

```python
data[0]
```

We can extract the data and save it to different variables:

```python
dt.datetime(2000,1,1)
```

```python
date = []
wind_x = []
wind_y = []
for i in range(len(data)):

    year = int(data[i][:4])
    month = int(data[i][4:6])
    day = int(data[i][6:8])
    date.append(dt.datetime(year, month, day))

    wind_x.append(np.float32(data[i][26:32]))
    wind_y.append(np.float32(data[i][33:36]))
wind_x = np.array(wind_x) # wind speed in x direction. eastward  = > 0 units are decimeters/sec
wind_y = np.array(wind_y) # wind speed in y direction. northward  = > 0 units are decimeters/sec

# Convert to m/s
wind_y = wind_y/10
wind_x = wind_x/10

# Calculate wind stress vector Tau_y = density_air (kg/m^3) x drag_coeff x |wind_speed| x wind_speed 
tau_y = 1.3*0.002*np.abs(wind_y)*wind_y # units are kg m-1 s-2
```

```python
plt.figure(figsize = (15, 5))
plt.plot(date, tau_y)
plt.xlabel('Date')
plt.ylabel('Wind stress in the y direction (kg m$^{-1}$ s$^{-2}$)')
plt.savefig('windstress_timeseries.jpg', bbox_inches = 'tight')
```

```python
# Calculate the offshore Ekman transport (in the x direction) using wind stress, the coriolis parameter, and average seawater density (1027 kg/m^3)

u_mass = tau_y / gsw.f(45) # mass transport kg/m/s
u_vol = u_mass/1027 # volume transport m3/m/sec
```

```python
plt.figure(figsize = (15, 5))
plt.plot(date, u_vol)
plt.xlabel('Date')
plt.ylabel('Ekman volume transport [m$^3$ m${^-1}$ s${^-1}$)')
plt.savefig('ekman_vol_trans_timeseries.jpg', bbox_inches = 'tight')
```

#### Programming challenge

Make a plot of the mass transport over time


Calculate and plot the values per month:

```python
tau_y_m = []
u_vol_m = []
for month in range(1, 13):
    index = []
    for d in range(len(date)):
        if date[d].month == month:
            index .append(d)
    tau_y_m.append(tau_y[index])
    u_vol_m.append(u_vol[index])
```

```python
fig, axs = plt.subplots(1, 2, figsize = (12, 5))
for i in range(12):
    axs[0].scatter(np.ones(len(tau_y_m[i]))+i, tau_y_m[i])
    axs[1].scatter(np.ones(len(u_vol_m[i]))+i, u_vol_m[i])
axs[0].set_ylabel('Wind stress in y direction (kg m$^{-1}$ s$^{-2}$)');
axs[1].set_ylabel('Ekman volume transport in x dir (m$^3$ m$^{-1}$ s$^{-1}$)');
fig.suptitle('Data grouped into month')
plt.savefig('monthly.jpg', bbox_inches = 'tight')
```

We can calculate the seasonal cycle:

```python
tau_y_sc = []
u_vol_sc = []
for month in range(1, 13):
    index = []
    for d in range(len(date)):
        if date[d].month == month:
            index .append(d)
    tau_y_sc.append(np.mean(tau_y[index]))
    u_vol_sc.append(np.mean(u_vol[index]))
tau_y_sc = np.array(tau_y_sc)
u_vol_sc = np.array(u_vol_sc)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (12, 5))
axs[0].plot(np.arange(1, 13, 1), tau_y_sc)
axs[0].set_ylabel('Wind stress in y direction (kg m$^{-1}$ s$^{-2}$)');

axs[1].plot(np.arange(1, 13, 1), u_vol_sc)
axs[1].set_ylabel('Ekman volume transport in x dir (m$^3$ m$^{-1}$ s$^{-1}$)');
fig.suptitle('Seasonal cycles')
plt.savefig('climatology.jpg', bbox_inches = 'tight')
```

A simple script to read in CTD data collected with the mini CTD (Castaway) used in the KSA205 field trip. Make sure this script and your CTD data file are in the same directory

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Replace `CC2225003_20221114_003455.csv` with the name of your file:

```python
data = pd.read_csv('CC2225003_20221114_003455.csv', skiprows = 28)
data
```

Let's extract the data and put it into variables for easier working.

```python
pressure = data.iloc[:,0]
depth = data.iloc[:,1] 
temperature = data.iloc[:,2]
salinity = data.iloc[:,5] 
density = data.iloc[:,7]
```

We can make a profile of salinity versus depth:

```python
plt.plot(salinity, depth, color = 'k', marker =  'o')
# Flip the y-axis to get surface at the top of the plot
plt.gca().invert_yaxis()
# Add labels
plt.xlabel('Salinity') # No units because it is sensor salinity
plt.ylabel('Depth (m)');
```

We can add the other two variables, temperature and density, in accompanying panels:

```python
fig, axs = plt.subplots(1, 3, figsize = (12, 4))
axs[0].plot(salinity, depth, color = 'k', marker = 'o')
axs[1].plot(temperature, depth, color = 'r', marker = 'o')
axs[2].plot(density, depth, color = 'b', marker = 'o')
# Invert axis:
for ax in axs:
    ax.invert_yaxis();
    # Add depth label
    ax.set_ylabel('Depth (m)')
# Add labels
axs[0].set_xlabel('Salinity')
axs[1].set_xlabel('Temperature ($^{\\circ}$C)')
axs[2].set_xlabel('Density (kg/m$^3$)');
# Save figure
plt.savefig('profile.jpg', bbox_inches = 'tight')
```
